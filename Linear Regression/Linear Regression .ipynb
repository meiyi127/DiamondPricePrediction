# 3. Data Preparation

## Check Missing Value

```python
# Check for missing values in each column
missing_values = df.isnull().sum()

print(missing_values)
total_missing_percentage = (diamonds.isnull().sum().sum() / (diamonds.shape[0] * diamonds.shape[1])) * 100

# Display the percentage of missing values
print(f"Percentage of missing values: {total_missing_percentage:.2f}%")
```

### Check Duplicate Records

```python
# Find duplicate records
duplicate_mask = df.duplicated()

# Create a DataFrame containing only the duplicate records
duplicate_rows = df[duplicate_mask]

print(duplicate_rows)
```

### Drop duplicate records

```python
# Drop duplicates
df = df.drop_duplicates()

# Calculate the percentage of duplicate rows after dropping them
duplicate_percentage_after_drop = (df.duplicated().sum() / len(df)) * 100

print(f"Percentage of Duplicate Rows After Dropping: {duplicate_percentage_after_drop:.2f}%")
```

### Data Normalization using z-score

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()

# Select the numerical columns you want to standardize
numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Apply the standardization to the selected columns
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Display the first few rows of the standardized DataFrame
print(df.head())
```

## Detect Outliers using IQR

```python
import numpy as np 

#IQR - detection for outliers 
def outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier = (data < lower_bound) | (data > upper_bound)
    return outlier 
```

```python
columns = ['carat', 'x', 'y', 'z', 'depth', 'table']

# Perform outlier detection for the specified columns
outlier = outliers_iqr(df[columns])

# Calculate the percentage of outliers for each column
percentage_outliers = (outlier.sum() / len(df)) * 100

# Display the rows with outliers
outlier_rows = df[outlier.any(axis=1)]
print(outlier_rows)

# Display the percentage of outliers for each column
for col, percentage in percentage_outliers.iteritems():
    print(f"Percentage of outliers in '{col}': {percentage:.2f}%")
```

### Replacing Outliers

```python
# Replace outliers with the median value
for column in columns:
    median_value = df[column].median()
    df[column] = df[column].where(~outlier[column], median_value)
```

## Data processing

```python
from sklearn.preprocessing import OrdinalEncoder

for col in df.columns:
    if df[col].dtype == "object":
        oe = OrdinalEncoder()
        oe.fit(df[[col]])
        df[col] = oe.fit_transform(df[[col]])
```

## Separating the data

```python
from sklearn.model_selection import train_test_split

X = df.drop('price', axis = 1)
y = df['price']

xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

# GridSearchCV for Parameter Setting


```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

# Define the parameter grid for GridSearchCV
param_grid = {
    'fit_intercept': [True, False],
    'copy_X': [True, False]
}

# Create a Linear Regression model
model = LinearRegression()

# Create the GridSearchCV object
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV object to the data
grid_search.fit(xtrain, ytrain)

# Print the best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score (Negative Mean Squared Error):", grid_search.best_score_)

# Get the best model from GridSearchCV
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
test_predictions = best_model.predict(xtest)
```

## Accuracy for Test and Training set

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predictions on the test set using the best model
test_predictions = best_model.predict(xtest)

# Calculate the Mean Squared Error (MSE) for the test set
mse_test = mean_squared_error(ytest, test_predictions)

# Calculate the Mean Absolute Error (MAE) for the test set
mae_test = mean_absolute_error(ytest, test_predictions)

# Calculate the R-squared (R^2) value for the test set
r2_test = r2_score(ytest, test_predictions)

# Calculate the Signal-to-Noise Ratio (SNR) for the test set
snr_test = 10 * np.log10(np.var(ytest) / mse_test)

# Print the calculated metrics
print("Mean Squared Error (MSE) for Testing Set:", f"{mse_test:.3f}")
print("Mean Absolute Error (MAE) for Testing Set:", f"{mae_test:.3f}")
print("R-squared (R^2) for Testing Set:", f"{r2_test:.3f}")
print("Signal-to-Noise Ratio (SNR) for Testing Set:", f"{snr_test:.3f} dB")
```
